### 3d Arrow code

```python
def arrow3d(ax, length=1, width=0.05, head=0.2, headwidth=2,
                theta_x=0, theta_z=0, offset=(0,0,0), **kw):
    w = width
    h = head
    hw = headwidth
    theta_x = np.deg2rad(theta_x)
    theta_z = np.deg2rad(theta_z)

    a = [[0,0],[w,0],[w,(1-h)*length],[hw*w,(1-h)*length],[0,length]]
    a = np.array(a)

    r, theta = np.meshgrid(a[:,0], np.linspace(0,2*np.pi,30))
    z = np.tile(a[:,1],r.shape[0]).reshape(r.shape)
    x = r*np.sin(theta)
    y = r*np.cos(theta)

    rot_x = np.array([[1,0,0],[0,np.cos(theta_x),-np.sin(theta_x) ],
                      [0,np.sin(theta_x) ,np.cos(theta_x) ]])
    rot_z = np.array([[np.cos(theta_z),-np.sin(theta_z),0 ],
                      [np.sin(theta_z) ,np.cos(theta_z),0 ],[0,0,1]])

    b1 = np.dot(rot_x, np.c_[x.flatten(),y.flatten(),z.flatten()].T)
    b2 = np.dot(rot_z, b1)
    b2 = b2.T+np.array(offset)
    x = b2[:,0].reshape(r.shape); 
    y = b2[:,1].reshape(r.shape); 
    z = b2[:,2].reshape(r.shape); 
    ax.plot_surface(x,y,z, **kw)
```

### fig_multi_var_1a

```python
plotname = "fig_multi_var_1a"
scale = 1
fig = plt.figure(figsize=(4*scale,4*scale))
ax = plt.axes(projection='3d')

# setting plot limits
scale = .75
xscale = .545
ax.set_xlim(-2.*scale*xscale, 2.*scale*xscale)
ax.set_ylim(-2.*scale*xscale, 2.*scale*xscale)
ax.set_zlim(-.2*scale, 1.75*scale)
# ax.axis('off')

# axis arrows
arrow3d(ax, length=5, width=0.004, head=0.05, headwidth=6, offset=[-2.25,0,0], 
        theta_x= 90, theta_z = 90,  color="black")

arrow3d(ax, length=5, width=0.004, head=0.05, headwidth=6, offset=[0,-2.25,0], 
        theta_x= -90, theta_z = 0,  color="black")


arrow3d(ax, length=1.5, width=0.013, head=0.05, headwidth=6, offset=[0,0,-.2], 
        theta_x= 0, theta_z = 90,  color="black")

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.view_init(26, 45)
ax.axis("off")

tickpointsize = 4
ticksize=7.5
ticks = [-2, 0.5, 1, 2]
zoffset = -.1
for tick in ticks:
    if np.abs(tick) == 2:
        ax.scatter3D(tick, 0, 0, s=tickpointsize, c='k')
        ax.text(tick, 0, zoffset, s=tick, fontsize=ticksize, ha="center", va="center")
        ax.scatter3D(0, tick, 0, s=tickpointsize, c='k')
        ax.text(0, tick, zoffset, s=tick, fontsize=ticksize, ha="center", va="center")
    if np.abs(tick) != 2:
        ax.scatter3D(0, 0, tick, s=tickpointsize, c='k')
        ax.text(0, .4, tick, s=tick, fontsize=ticksize, ha="center", va="center")

ax.text(2.9, 0, 0, s="x", fontsize=ticksize, ha="center", va="center")
ax.text(0, 2.9, 0, s="y", fontsize=ticksize, ha="center", va="center")
ax.text(0, 0, 1.35, s="z", fontsize=ticksize, ha="center", va="center")

ax.set_box_aspect([1,1,4])

# Plot
N = 5
max = 2.4
x = np.linspace(-max, max, N)
y = np.linspace(-max, max, N)
X, Y = np.meshgrid(x, y)
z = lambda x,y: 1 / np.sqrt(1 + x**2 + y**2)
ax.scatter3D(X, Y, z(X,Y), color='k', s=10)

fig.savefig(f"plots/{plotname}.pdf", pad_inches=0.01, transparent="True", bbox_inches='tight')
```
