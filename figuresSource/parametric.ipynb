# Hoofdstuk 11 Poolcoördinaten en parametervergelijkingen

```python
from mpl_toolkits.axisartist.axislines import SubplotZero
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib
import numpy as np
matplotlib.rcParams["font.family"] = "monospace"
matplotlib.rcParams['font.style'] = "italic"
plt.rcParams.update({'mathtext.default':  'regular' })
```

## Poolcoördinaten

### fig_parametric_1

```python
fig1 = plt.figure(figsize=(4,4))
ax1 = fig1.add_axes([0,0,1,1])
ax1.set_xlim(-5.1, 5.1)
ax1.set_ylim(-5.1, 5.1)

ax1.set_xticks([])
ax1.set_yticks([])

for direction in ["left", "right", "bottom", "top"]:
    # hides borders
    ax1.spines[direction].set_color('none')
    if direction in ["left", "bottom"]:
        ax1.spines[direction].set_position('zero')

ax1.arrow(0, 0, -3, 4., fc='k', ec='k', lw = .5, head_width=0.3, head_length=0.3, overhang = 0.3, length_includes_head= True, clip_on = False)
ax1.arrow(0, 0, 5, 0, fc='k', ec='k', lw = .5, head_width=0.3, head_length=0.3, overhang = 0.3, length_includes_head= True, clip_on = False)

ax1.scatter([0, -3], [0, 4], color='black', zorder=10)

ax1.text(1.6, 2, '\u03B8', color="black", size=15)
ax1.text(-0.5, -1, 'Pole', color="black", size=15)
ax1.text(5.1, 0, 'Polar axis', color="black", size=15)
ax1.text(-2, 1, 'r', color="black", size=15)
ax1.text(2, -1, 'r', color="black", size=15)
ax1.text(-6, 4, 'P(r,\u03B8)', color="black", size=15)

path = matplotlib.path.Path(np.array([np.array([2*np.cos(theta), 2*np.sin(theta)]) for theta in np.linspace(0, np.arctan(-3/4)+np.pi-0.3, 100)]), codes=None, _interpolation_steps=1, closed=False, readonly=False)
arc = patches.FancyArrowPatch(path=path, arrowstyle="-|>, head_width=2, head_length=2", color="k", lw=2)

circle = plt.Circle((0, 0), 5, color='black', fill=False, zorder=0)
ax1.add_patch(circle)
for r in range(0, 5):
    circle = plt.Circle((0, 0), r, color='grey', fill=False, zorder=0)
    ax1.add_patch(circle)
    
for theta in np.arange(0, 2*np.pi, np.pi/12):
    ax1.plot([0, 5*np.cos(theta)], [0, 5*np.sin(theta)], color='grey', lw=1, zorder=0)

ax1.add_patch(arc)

fig1.savefig("fig_parametric_1.pdf",bbox_inches='tight')
plt.show()
```

### fig_parametric_8b

```python
def draw_brace(ax, xspan, theta):
    """Draws an annotated brace on the axes."""
    xmin, xmax = xspan
    xspan = xmax - xmin
    ax_xmin, ax_xmax = ax.get_xlim()
    xax_span = ax_xmax - ax_xmin
    ymin, ymax = ax.get_ylim()
    yspan = ymax - ymin
    resolution = int(xspan/xax_span*100)*2+1 # guaranteed uneven
    beta = 300./xax_span # the higher this is, the smaller the radius

    x = np.linspace(xmin, xmax, resolution)
    x_half = x[:resolution//2+1]
    y_half_brace = (1/(1.+np.exp(-beta*(x_half-x_half[0])))
                    + 1/(1.+np.exp(-beta*(x_half-x_half[-1]))))
    y = np.concatenate((y_half_brace, y_half_brace[-2::-1]))
    y = (.05*y - .01)*yspan # adjust vertical position

    ax.autoscale(False)
    ax.plot(x*np.cos(theta) - y*np.sin(theta), x*np.sin(theta) + y*np.cos(theta), color='black', lw=1)
```
