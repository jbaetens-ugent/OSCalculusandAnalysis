### fig_lim_7

```python
plotname = "fig_lim_7"
fig = plt.figure(figsize=(4.5, 3))
ax = fig.add_axes([0, 0, 1, 1])

def draw_brace(ax, xspan, theta, c=0, d=0):
    """Draws an annotated brace on the axes."""
    xmin, xmax = xspan
    xspan = xmax - xmin
    ax_xmin, ax_xmax = ax.get_xlim()
    xax_span = ax_xmax - ax_xmin
    ymin, ymax = ax.get_ylim()
    yspan = ymax - ymin
    resolution = int(xspan/xax_span*100)*2+1 # guaranteed uneven
    beta = 300./xax_span # the higher this is, the smaller the radius

    x = np.linspace(xmin, xmax, resolution)
    x_half = x[:resolution//2+1]
    y_half_brace = (1/(1.+np.exp(-beta*(x_half-x_half[0])))
                    + 1/(1.+np.exp(-beta*(x_half-x_half[-1]))))
    y = np.concatenate((y_half_brace, y_half_brace[-2::-1]))
    y = (.05*y - .01)*yspan # adjust vertical position

    ax.autoscale(False)
    ax.plot(x*np.cos(theta) - y*np.sin(theta)+c, x*np.sin(theta) + y*np.cos(theta)+d, color='black', lw=1)

# Opmaak
for direction in ["left", "right", "bottom", "top"]:
    # hides borders
    ax.spines[direction].set_color('none')
    if direction in ["left", "bottom"]:
        ax.spines[direction].set_position('zero')

ax.set_xlim(-.2, 2.5)
ax.set_ylim(-.2, np.sqrt(3))

ax.arrow(0, 0, 2.3, 0, fc='k', ec='k', lw = .6, head_width=0.08, head_length=0.12, overhang = 0.3, length_includes_head= True, clip_on = False)
ax.arrow(0, 0, 0, 1.6, fc='k', ec='k', lw = .6, head_width=0.08, head_length=.12, overhang = 0.3, length_includes_head= True, clip_on = False)

# ax.axis('off')
ax.text(2.4, 0, s="$x$", fontsize=15, ha="center", va="center")
ax.text(0, 1.7, s="$y$", fontsize=15, ha="center", va="center")

# Plot
x1 = np.linspace(0, 2.3, 1000)
ax.plot(x1, np.sqrt(x1), c='k', lw=1.5)

c=1.2
L=np.sqrt(c)
ax.scatter(c, np.sqrt(c), c='k', s=40, zorder=10)

epsilon = .25
dotted_linewidth = 1.5
ax.plot([0, (L-epsilon)**2], [L-epsilon, L-epsilon], c='k', lw=dotted_linewidth, ls='--')
ax.plot([0, (L+epsilon)**2], [L+epsilon, L+epsilon], c='k', lw=dotted_linewidth, ls='--')
ax.plot([(L-epsilon)**2, (L-epsilon)**2], [0, L-epsilon], c='k', lw=dotted_linewidth, ls='--')
ax.plot([(L+epsilon)**2, (L+epsilon)**2], [0, L+epsilon], c='k', lw=dotted_linewidth, ls='--')

ax.set_xticks([(L+epsilon)**2 - L**2, L**2, (L+epsilon)**2])
ax.set_yticks([L-epsilon, L, L+epsilon])
ax.set_xticklabels([r"$c - \delta$", "c", r"$c + \delta$"])
ax.set_yticklabels([r"$L - \varepsilon$", "L", r"$L + \varepsilon$"])

draw_brace(ax, [L-epsilon, L], -np.pi/2, c=0, d=L-epsilon + 1.1)
draw_brace(ax, [L, L+epsilon], -np.pi/2, c=0, d=L-epsilon + 1.6)
draw_brace(ax, [L**2, (L+epsilon)**2], 0, c=0, d=0)
draw_brace(ax, [(L-epsilon)**2, L**2], 0, c=0, d=0)

ax.text(((L+epsilon)**2 + L**2)/2, .2, s=r"$\delta$", fontsize=15, ha="center", va="center")
ax.text(((L-epsilon)**2 + L**2)/2, .2, s=r"$\delta$", fontsize=15, ha="center", va="center")
ax.text(.2, (L-epsilon + L)/2, s=r"$\varepsilon$", fontsize=15, ha="center", va="center")
ax.text(.2, (L+epsilon + L)/2, s=r"$\varepsilon$", fontsize=15, ha="center", va="center")

fig.savefig(f"plots/{plotname}.pdf", pad_inches=0.01, transparent="True", bbox_inches='tight')
```

```python
plotname = "fig_lim_8b"
fig = plt.figure(figsize=(4, 3))

ax = fig.add_axes([0, 0, 1, 1])

def draw_brace(ax, xspan, theta, c=0, d=0):
    """Draws an annotated brace on the axes."""
    xmin, xmax = xspan
    xspan = xmax - xmin
    ax_xmin, ax_xmax = ax.get_xlim()
    xax_span = ax_xmax - ax_xmin
    ymin, ymax = ax.get_ylim()
    yspan = ymax - ymin
    resolution = int(xspan/xax_span*100)*2+1 # guaranteed uneven
    beta = 300./xax_span # the higher this is, the smaller the radius

    x = np.linspace(xmin, xmax, resolution)
    x_half = x[:resolution//2+1]
    y_half_brace = (1/(1.+np.exp(-beta*(x_half-x_half[0])))
                    + 1/(1.+np.exp(-beta*(x_half-x_half[-1]))))
    y = np.concatenate((y_half_brace, y_half_brace[-2::-1]))
    y = (.05*y - .01)*yspan # adjust vertical position

    ax.autoscale(False)
    ax.plot(x*np.cos(theta) - y*np.sin(theta)+c, x*np.sin(theta) + y*np.cos(theta)+d, color='black', lw=1)

# Opmaak
for direction in ["left", "right", "bottom", "top"]:
    # hides borders
    ax.spines[direction].set_color('none')
    if direction in ["left", "bottom"]:
        ax.spines[direction].set_position('zero')

ax.set_xlim(-.2, 8)
ax.set_ylim(-.2, 3.5)

ax.arrow(0, 0, 7, 0, fc='k', ec='k', lw = .6, head_width=0.1, head_length=0.25, overhang = 0.3, length_includes_head= True, clip_on = False)
ax.arrow(0, 0, 0, 3, fc='k', ec='k', lw = .6, head_width=0.14, head_length=.18, overhang = 0.3, length_includes_head= True, clip_on = False)

# ax.axis('off')
ax.text(7.2, 0, s="$x$", fontsize=15, ha="center", va="center")
ax.text(0, 3.2, s="$y$", fontsize=15, ha="center", va="center")

# Plot
x1 = np.linspace(0, 7, 1000)
ax.plot(x1, np.sqrt(x1), c='k', lw=1.5)

c=4
L=2
ax.scatter(c, np.sqrt(c), c='k', s=40, zorder=10)

epsilon = .5
dotted_linewidth = 1.5
ax.plot([0, (L-epsilon)**2], [L-epsilon, L-epsilon], c='k', lw=dotted_linewidth, ls='--')
ax.plot([0, (L+epsilon)**2], [L+epsilon, L+epsilon], c='k', lw=dotted_linewidth, ls='--')
ax.plot([(L-epsilon)**2, (L-epsilon)**2], [0, L-epsilon], c='k', lw=dotted_linewidth, ls='--')
ax.plot([(L+epsilon)**2, (L+epsilon)**2], [0, L+epsilon], c='k', lw=dotted_linewidth, ls='--')

ax.set_xticks([2, 4, 6])
ax.set_yticks([1, 2])

draw_brace(ax, [L-epsilon, L], -np.pi/2, c=0.1, d=4)
draw_brace(ax, [L, L+epsilon], -np.pi/2, c=0.1, d=4)
draw_brace(ax, [L**2, (L+epsilon)**2], 0, c=0, d=0)
draw_brace(ax, [(L-epsilon)**2, L**2], 0, c=0, d=0)

ax.text(1.5, (L-epsilon + L)/2, s=r"$\varepsilon=0.5$", fontsize=11, ha="center", va="center")
ax.text(1.5, (L+epsilon + L)/2, s=r"$\varepsilon=0.5$", fontsize=11, ha="center", va="center")
ax.text(((L+epsilon)**2 + L**2)/2 + .1, .4, s="width = 1.75", fontsize=8, ha="center", va="center")
ax.text(((L-epsilon)**2 + L**2)/2 + .1, .4, s="width = 2.25", fontsize=8, ha="center", va="center")

ax.text(4.2, 1.2, s=r"... choose $\delta$ smaller" , fontsize=11, ha="center", va="center")
ax.text(4.2, .9, s="than each of these", fontsize=11, ha="center", va="center")

fig.savefig(f"plots/{plotname}.pdf", pad_inches=0.01, transparent="True", bbox_inches='tight')
```

```python
plotname = "fig_lim_17a"
fig = plt.figure(figsize=(4, 5.2))
ax = fig.add_axes([0, 0, 1, 1])

def draw_brace(ax, xspan, theta, c=0, d=0):
    """Draws an annotated brace on the axes."""
    xmin, xmax = xspan
    xspan = xmax - xmin
    ax_xmin, ax_xmax = ax.get_xlim()
    xax_span = ax_xmax - ax_xmin
    ymin, ymax = ax.get_ylim()
    yspan = ymax - ymin
    resolution = int(xspan/xax_span*100)*2+1 # guaranteed uneven
    beta = 300./xax_span # the higher this is, the smaller the radius

    x = np.linspace(xmin, xmax, resolution)
    x_half = x[:resolution//2+1]
    y_half_brace = (1/(1.+np.exp(-beta*(x_half-x_half[0])))
                    + 1/(1.+np.exp(-beta*(x_half-x_half[-1]))))
    y = np.concatenate((y_half_brace, y_half_brace[-2::-1]))
    y = (.05*y - .01)*yspan * .4 # adjust vertical position

    ax.autoscale(False)
    ax.plot(x*np.cos(theta) - y*np.sin(theta)+c, x*np.sin(theta) + y*np.cos(theta)+d, color='black', lw=1.5)

# Opmaak
for direction in ["left", "right", "bottom", "top"]:
    # hides borders
    ax.spines[direction].set_color('none')
    if direction in ["left", "bottom"]:
        ax.spines[direction].set_position('zero')

ax.set_xlim(-.4, 2)
ax.set_ylim(-.1, 3)

ax.arrow(-.1, 0, 1.9, 0, fc='k', ec='k', lw = .8, head_width=0.06, head_length=0.1, overhang = 0.3, length_includes_head= True, clip_on = False)
ax.arrow(0, -.1, 0, 2.9, fc='k', ec='k', lw = .8, head_width=0.06, head_length=0.1, overhang = 0.3, length_includes_head= True, clip_on = False)

# ax.axis('off')
ax.text(1.9, 0, s="$x$", fontsize=15, ha="center", va="center")
ax.text(0, 2.9, s="$y$", fontsize=15, ha="center", va="center")

# Plot
x = np.linspace(0, 1.5, 1000)
f = lambda x : (x+.15)**2-.022
ax.plot(x, f(x), c='k', lw=1.5, zorder=0)

delta_outer = .4
c = 1
a = c - delta_outer
b = c + delta_outer
delta = .2
ax.plot([a, a], [0, f(a)], c='gray', lw=1.2, zorder=0)
ax.plot([b, b], [0, f(b)], c='gray', lw=1.22, zorder=0)
ax.plot([0, a], [f(a), f(a)], c='gray', lw=1.2, zorder=0)
ax.plot([0, b], [f(b), f(b)], c='gray', lw=1.2, zorder=0)

ax.plot([c, c], [0, f(c)], ls='dotted', c='k', lw=1.2, zorder=0)
ax.plot([0, c], [f(c), f(c)], ls='dotted', c='k', lw=1.2, zorder=0)

ax.plot([c + delta, c + delta], [0, f(c + delta)], '--', c='k', lw=1.2, zorder=0)
ax.plot([0, c + delta], [f(c + delta), f(c + delta)], '--', c='k', lw=1.2, zorder=0)

ax.plot([c - delta, c - delta], [0, f(c - delta)], '--', c='k', lw=1.2, zorder=0)
ax.plot([0, c - delta], [f(c - delta), f(c - delta)], '--', c='k', lw=1.2, zorder=0)

ax.scatter(a, f(a), c='k', s=30, zorder=1)
ax.scatter(b, f(b), c='k', s=30, zorder=1)

draw_brace(ax, [c-delta, c], 0, 0, .1)
draw_brace(ax, [f(c-delta), f(c)], -np.pi/2, 0, 2.18)

ax.set_xticks([a, c, b])
ax.set_yticks([f(a), f(c), f(b)])
ax.set_xticklabels(["a", "c", "b"])
ax.set_yticklabels(["f(a)", "u=f(c)?", "f(b)"])

ax.text((c - delta + c)/2, .25, s="$\delta$", fontsize=15, ha="center", va="center")
ax.text(.15, (f(c - delta) + f(c))/2, s=r"$\varepsilon$", fontsize=15, ha="center", va="center")

fig.savefig(f"plots/{plotname}.pdf", pad_inches=0.01, transparent="True", bbox_inches='tight')
```
