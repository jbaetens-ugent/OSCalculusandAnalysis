### fig_vector_3a

```python
plotname = "fig_vector_3a"

import matplotlib.patches as patches

fig = plt.figure(figsize=(4, 5))
ax = fig.add_axes([0, 0, 1, 1])

# Opmaak
xmax, ymax = 3, 5
xmin, ymin = -.5, -1
delta = .05
ax.set_xlim(xmin, xmax)
ax.set_ylim(ymin, ymax)

# Plot
ax.arrow(xmin, 0, xmax - xmin - delta, 0, fc='k', ec='k', lw = .8, head_width=0.12, head_length=0.12, overhang = 0.3, length_includes_head= True, clip_on = True)
ax.arrow(0, ymin, 0, ymax - ymin - delta, fc='k', ec='k', lw = .8, head_width=0.08, head_length=0.2, overhang = 0.3, length_includes_head= True, clip_on = True)

Vlength = 4
Wlength = .8
Vangle = .8*np.pi/5
Wangle = np.pi/4
V = np.array((Vlength*np.sin(Vangle), Vlength*np.cos(Vangle)))
W = np.array((Wlength*np.sin(Wangle), -Wlength*np.cos(Wangle)))
VplusW = V + W

ax.arrow(0, 0, V[0], V[1], fc='k', ec='k', lw = 1.2, head_width=0.06, head_length=0.12, overhang = 0.3, length_includes_head= True, clip_on = True)
ax.arrow(0, 0, W[0], W[1], fc='k', ec='k', lw = 1.2, head_width=0.07, head_length=0.08, overhang = 0.3, length_includes_head= True, clip_on = True)
ax.arrow(.98*VplusW[0], .98*VplusW[1], .02*VplusW[0], .02*VplusW[1], fc='k', ec='k', lw = 1.2, head_width=0.06, head_length=0.12, overhang = 0.3, length_includes_head= True, clip_on = True)

ax.plot((W[0], VplusW[0]), (W[1], VplusW[1]), "--", c='k', lw=1)
ax.plot((V[0], VplusW[0]), (V[1], VplusW[1]), "--", c='k', lw=1)
ax.plot((0, VplusW[0]), (0, VplusW[1]), "--", c='k', lw=1)

style = "Simple, tail_width=0.3, head_width=3, head_length=6"
kw = dict(arrowstyle=style, color="k")

a1 = patches.FancyArrowPatch((.1, ymax/2), (1, .8*ymax/2), connectionstyle="arc3,rad=-.15", **kw)
a2 = patches.FancyArrowPatch((.04, 1.2*ymin/2), (.3, .35*ymin), connectionstyle="arc3,rad=.15", **kw)
ax.add_patch(a1)
ax.add_patch(a2)

textsize = 10
ax.text(V[0], V[1] + .1, s=r"$\vec{V}$", size=textsize)
ax.text(VplusW[0], VplusW[1] + .1, s=r"$\vec{V}+\vec{W}$", size=textsize)
ax.text(W[0], W[1] - .2, s=r"$\vec{W}$", size=textsize)

ax.text(.4, ymax/2, s=r"$40^{\circ}$")
ax.text(.2, .8*ymin, s=r"$60^{\circ}$")

ax.text(xmax, -.05, s="E", size=15)
ax.text(-.05, ymax, s="N", size=15)
ax.axis("off")
fig.savefig(f"plots/{plotname}.pdf", pad_inches=0.1, transparent="True", bbox_inches='tight')
```

### fig_vector_8

```python
plotname = "fig_vector_13"

fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')

def arrow3d(ax, length=1, width=0.05, head=0.2, headwidth=2,
                theta_x=0, theta_z=0, offset=(0,0,0), **kw):
    w = width
    h = head
    hw = headwidth
    theta_x = np.deg2rad(theta_x)
    theta_z = np.deg2rad(theta_z)

    a = [[0,0],[w,0],[w,(1-h)*length],[hw*w,(1-h)*length],[0,length]]
    a = np.array(a)

    r, theta = np.meshgrid(a[:,0], np.linspace(0,2*np.pi,30))
    z = np.tile(a[:,1],r.shape[0]).reshape(r.shape)
    x = r*np.sin(theta)
    y = r*np.cos(theta)

    rot_x = np.array([[1,0,0],[0,np.cos(theta_x),-np.sin(theta_x) ],
                      [0,np.sin(theta_x) ,np.cos(theta_x) ]])
    rot_z = np.array([[np.cos(theta_z),-np.sin(theta_z),0 ],
                      [np.sin(theta_z) ,np.cos(theta_z),0 ],[0,0,1]])

    b1 = np.dot(rot_x, np.c_[x.flatten(),y.flatten(),z.flatten()].T)
    b2 = np.dot(rot_z, b1)
    b2 = b2.T+np.array(offset)
    x = b2[:,0].reshape(r.shape); 
    y = b2[:,1].reshape(r.shape); 
    z = b2[:,2].reshape(r.shape); 
    ax.plot_surface(x,y,z, **kw)

# Opmaak
ax.set_axis_off()
ax.view_init(30, 65)

labelpos_x = [1, 2, 3]
labelpos_y = [1, 2]
labelpos_z = [1, 2, 3]

ticksize = 8
for x in labelpos_x:
    ax.scatter(x, 0, 0, c='k', s=ticksize)

for y in labelpos_y:
    ax.scatter(0, y, 0, c='k', s=ticksize)

for z in labelpos_z:
    ax.scatter(0, 0, z, c='k', s=ticksize)

tickmarksize = 15
for x in labelpos_x:
    ax.text(x, 0, -.5, s=str(x), size=tickmarksize)

for y in labelpos_y:
    ax.text(-.25, y, 0, s=str(y), size=tickmarksize)

for z in labelpos_z:
    ax.text(0, .25, z, s=str(z), size=tickmarksize)

ax.scatter(0, 0, 0, c='k', s=ticksize)
ax.text(.25, .25, -.25, s="0", size=tickmarksize)

# Plot
scale = .85
ax.set_xlim(-.25*scale, 4*scale)
ax.set_ylim(-.25*scale, 3*scale)
ax.set_zlim(-.25*scale, 4*scale)

arrow3d(ax, length=3.25, width=0.01, head=0.05, headwidth=6, offset=[0,-.25,0], 
        theta_x= -90, theta_z = 0,  color="black")

arrow3d(ax, length=4.25, width=0.01, head=0.05, headwidth=6, offset=[-.25,0,0], 
        theta_x= 90, theta_z = 90,  color="black")

arrow3d(ax, length=4.25, width=0.01, head=0.05, headwidth=6, offset=[0,0,-.25], 
        theta_x= 0, theta_z = 90,  color="black")

linewidth = 1
plt.plot([2,2], [0,1], [0,0], '--', c='gray', lw=linewidth)
plt.plot([0,2], [1,1], [0,0], '--', c='gray', lw=linewidth)
plt.plot([2,2], [0,0], [0,3], '--', c='gray', lw=linewidth)
plt.plot([2,2], [1,1], [0,3], '--', c='gray', lw=linewidth)
plt.plot([0,0], [1,1], [0,3], '--', c='gray', lw=linewidth)
plt.plot([0,2], [0,0], [3,3], '--', c='gray', lw=linewidth)
plt.plot([2,2], [0,1], [3,3], '--', c='gray', lw=linewidth)
plt.plot([0,2], [1,1], [3,3], '--', c='gray', lw=linewidth)
plt.plot([0,0], [0,1], [3,3], '--', c='gray', lw=linewidth)

textsize = 18

P = np.array((2, 1, 3))
ax.scatter(P[0], P[1], P[2], c='k', s=40)
ax.text(P[0], P[1], P[2] + .1, s="P", size=textsize)

ax.text(4, 0, .25, s="x", size=textsize)
ax.text(0, 3, .25, s="y", size=textsize)
ax.text(0, .25, 4, s="z", size=textsize)

fig.savefig(f"plots/{plotname}.pdf", pad_inches=0, transparent="True", bbox_inches='tight')
```
