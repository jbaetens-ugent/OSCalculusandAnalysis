### fig_int_1a

```python
plotname = "fig_int_1a"
scale=.7
fig = plt.figure(figsize=(6.5*scale, 4*scale))
ax = fig.add_axes([0, 0, 1, 1])

# Opmaak
for direction in ["left", "right", "bottom", "top"]:
    # hides borders
    ax.spines[direction].set_color('none')
    if direction in ["left", "bottom"]:
        ax.spines[direction].set_position('zero')

ax.set_xlim(-.5, 12)
ax.set_ylim(-.1, 6)

ax.set_xticks([5, 10])
ax.set_yticks([5])

ax.arrow(0, 0, 10.75, 0, fc='k', ec='k', lw = .7, head_width=0.19, head_length=0.45, overhang = 0.3, length_includes_head= True, clip_on = False, zorder=0)
ax.arrow(0, 0, 0, 5.5, fc='k', ec='k', lw = .7, head_width=0.22, head_length=0.32, overhang = 0.3, length_includes_head= True, clip_on = False, zorder=0)

ax.text(11.5, 0, s="$t(s)$", fontsize=12, ha="center", va="center")
ax.text(0, 5.75, s="y(m/s)", fontsize=12, ha="center", va="center")

# Plot
ax.plot([0, 10], [5, 5], color="k", lw=1.3)
ax.fill_between([0, 10], [0, 0], [5, 5], color="lightgray", zorder=-1)
fig.savefig(f"plots/{plotname}.pdf", pad_inches=0.01, transparent="True", bbox_inches='tight')
```
