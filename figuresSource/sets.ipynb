### fig_sets_1

```python
plotname = "fig_sets_1"

from matplotlib.patches import Ellipse
from matplotlib.patches import Rectangle

fig = plt.figure(figsize=(4.5, 4))
ax = fig.add_axes([0, 0, 1, 1])

# Opmaak
ax.axis('off')
ax.set_xlim(-1.05, 1.25)
ax.set_ylim(-1.05, 1.05)

# Plot
edge = 0.02
circle1 = Ellipse((0, 0), 2, 2, color='k')
circle2 = Ellipse((0, 0), 2 - edge, 2 - edge, color='white')
ax.add_patch(circle1)
ax.add_patch(circle2)

rectangle = Rectangle((1.06, -.1), 0.14, 0.2, ec='k', fill=False)
ax.add_patch(rectangle)

ax.plot((1, 1.05), (0, 0), c='k')

ax.text(-.6, 0, s="k",  c="cornflowerblue", style = "italic", size=15)
ax.text(.6, 0, s="l",  c="gold", style = "italic", size=15)
ax.text(0, 0, s="m",  c="olivedrab", style = "italic", size=15)
ax.text(0, .6, s="s",  c="slateblue", style = "italic", size=15)
ax.text(0, -.6, s="o",  c="orangered", style = "italic", size=15)

ax.text(1.1, -.04, s="S",  c="k", style = "italic", size=15)

fig.tight_layout()
fig.savefig(f"plots/{plotname}.pdf", pad_inches=0.05, transparent="True", bbox_inches='tight')
```

### fig_sets_4

```python
plotname = "fig_sets_4"

import matplotlib_venn as vplt

fig = plt.figure(figsize=(4, 4))
ax = fig.add_axes([0, 0, 1, 1])

# Opmaak
E = set("pigeon bike Afrikaans bike computer internet fest blitzkrieg".split(" "))
D = set("Afrikaans fiets duif computer internet berhaupt".split(" "))
G = set("blitzkrieg fest computer internet berhaupt rad taube".split(" "))

v = vplt.venn3_unweighted([E, D, G], ("E", "D", "G"))

for id in ('100', '010', '001', '110', '011', '101', '111'):
    v.get_patch_by_id(id).set_color('white')

for idx, subset in enumerate(v.subset_labels):
    v.subset_labels[idx].set_visible(False)

vplt.venn3_circles(subsets = (1, 1, 1, 1, 1, 1, 1), linewidth=.5)

color1 = "blue"
color2 = "orange"

ax.text(-0.1, 0.35, s="Afrikaans", c=color1, size=8)
ax.text(-0.4, 0.38, s="pigeon", c=color2, size=8)
ax.text(-0.45, 0.25, s="bike", c=color1, size=8)
ax.text(-0.36, -0.25, s="blitzkrieg", c=color1, size=8)
ax.text(-0.35, -0.1, s="fest", c=color2, size=8)
ax.text(-0.1, 0, s="computer", c=color1, size=8)
ax.text(-0.1, -0.1, s="internet", c=color2, size=8)
ax.text(0.15, -0.25, s="berhaupt", c=color1, size=8)
ax.text(-0.2, -0.4, s="rad", c=color1, size=8)
ax.text(-0.03, -0.5, s="taube", c=color2, size=8)
ax.text(0.35, 0.28, s="fiets", c=color2, size=8)
ax.text(0.44, 0.15, s="duif", c=color1, size=8)


fig.savefig(f"plots/{plotname}.pdf", pad_inches=0.05, transparent="True", bbox_inches='tight')
```
