### fig_behaviour_1a

```python
plotname = "fig_behaviour_1a"
scale=1.2
fig = plt.figure(figsize=(4*scale, 4.2*scale))
ax = fig.add_axes([0, 0, 1, 1])

# Opmaak
for direction in ["left", "right", "bottom", "top"]:
    # hides borders
    ax.spines[direction].set_color('none')
    if direction in ["left", "bottom"]:
        ax.spines[direction].set_position('zero')

ax.set_xlim(-2.2, 2.2)
ax.set_ylim(-.5, 6.5)

ax.set_xticks([-2, -1 ,1, 2])
ax.set_yticks([2, 4])

ax.arrow(-2.2, 0, 4.4, 0, fc='k', ec='k', lw = .8, head_width=0.1, head_length=0.1, overhang = 0.3, length_includes_head= True, clip_on = False)
ax.arrow(0, -.5, 0, 6.5, fc='k', ec='k', lw = .8, head_width=0.06, head_length=0.18, overhang = 0.3, length_includes_head= True, clip_on = False)

ax.text(2.25, 0, s="$x$", fontsize=15, ha="center", va="center")
ax.text(0, 6.2, s="$y$", fontsize=15, ha="center", va="center")

# Plot
x = np.linspace(-2, 2, 100)

ax.plot(x, -x**2 + 5, c='k', lw=1.8, zorder=0)

pointsize = 70
ax.scatter(-2, 1, c='w', edgecolors='k', s=pointsize, zorder=1)
ax.scatter(2, 1, c='w', edgecolors='k', s=pointsize, zorder=1)

ax.annotate('', xy=(-2.02, 0), xytext=(2.02, 0), arrowprops=dict(arrowstyle="]-[", color='k', lw=1.3), color='k', fontsize=6, ha="center", va="center")
ax.plot([-1.97, 1.97], [0, 0], c='k', lw=3, zorder=0)
fig.savefig(f"plots/{plotname}.pdf", pad_inches=0.01, transparent="True", bbox_inches='tight')
```
