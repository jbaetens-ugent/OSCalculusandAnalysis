### FreeFall_1

```python
figFreeFall1 = plt.figure(figsize=(6,4))
axFreeFall1 = figFreeFall1.add_axes([0,0,1,1])
axFreeFall1.set_xlim(0, 30)
axFreeFall1.set_ylim(0, 250)

import sympy as sp

lines = ['-', '--', ':', '-.', '-']
col = ['black', 'black', 'black', 'black', 'grey']
for i, initial in enumerate([0, 60, 120, 180, 240]):
    t = sp.symbols("t")
    v = sp.Function('v')
    r=5.25
    m=75
    eq = sp.Eq(9.81-r*v(t)/m, sp.diff(v(t), t))
    diff = sp.solvers.ode.dsolve(eq, v(t), ics={v(0): initial})
    lam_t = sp.lambdify(t, diff.rhs)
    t = np.linspace(0, 30, 200)
    axFreeFall1.plot(t, lam_t(t), ls=lines[i], color=col[i], label='v$_0$ = {}'.format(initial))

axFreeFall1.set_xlabel('Time (s)')
axFreeFall1.set_ylabel('Velocity (m/s)')
axFreeFall1.legend()
figFreeFall1.savefig("FreeFall_1.pdf",bbox_inches='tight')
plt.show()
```

### Freefall_Math1

```python
figFreefallM1 = plt.figure(figsize=(6,4))
axFreefallM1 = figFreefallM1.add_axes([0,0,1,1])
axFreefallM1.set_xlim(0, 30)
axFreefallM1.set_ylim(0, 125)
for direction in ["right", "top"]:
    # hides borders
    axFreefallM1.spines[direction].set_color('none')
    if direction in ["left", "bottom"]:
        axFreefallM1.spines[direction].set_position('zero')

import sympy as sp

t = sp.symbols("t")
v = sp.Function('v')
r=5.25
m=75
eq = sp.Eq(9.81-r*v(t)/m, sp.diff(v(t), t))
diff = sp.solvers.ode.dsolve(eq, v(t), ics={v(0): 0})
lam_t = sp.lambdify(t, diff.rhs)
t = np.linspace(0, 30, 200)
axFreefallM1.plot(t, lam_t(t), color='black')

axFreefallM1.set_xlabel('Time (s)')
axFreefallM1.set_ylabel('Velocity (m/s)')
figFreefallM1.savefig("Freefall_Math2.pdf",bbox_inches='tight')
plt.show()
```

### Malthus_VB

```python
figMalthus_VB = plt.figure(figsize=(6,4))
axMalthus_VB = figMalthus_VB.add_axes([0,0,1,1])
axMalthus_VB.set_xlim(0, 15)
axMalthus_VB.set_ylim(0, 900)

from sympy import Function, rsolve, symbols
k = symbols('k')
N = Function("N")
f = N(k+1) - 1.5*N(k)
sol = rsolve(f, N(k), {N(0):2})
lam = sp.lambdify(k, sol)
for i in range(0, 16):
    axMalthus_VB.scatter([i], [lam(i)], color='black')
    axMalthus_VB.plot([i, i], [0, lam(i)], color='black')

axMalthus_VB.set_xlabel('Generation k')
axMalthus_VB.set_ylabel('Population size N$_k$')
figMalthus_VB.savefig("Malthus_VB.pdf",bbox_inches='tight')
plt.show()
```

### oplossingE

```python
figoplossingE = plt.figure(figsize=(6,4))
axoplossingE = figoplossingE.add_axes([0,0,1,1])
axoplossingE.set_xlim(0, 5)
axoplossingE.set_ylim(-0.5, 1)

import sympy as sp
x = sp.symbols("x")
y = sp.Function('y')
eq = sp.Eq(sp.diff(sp.diff(y(x), x), x) + 2*sp.diff(y(x), x) + 5*y(x), 0)
diff = sp.solvers.ode.dsolve(eq, y(x), ics={y(0): 2, y(x).diff(x).subs(x, 0): 2})
lam_x = sp.lambdify(x, diff.rhs)
x = np.linspace(0, 5, 100)
axoplossingE.plot(x, lam_x(x), color='black')

axoplossingE.set_xlabel('x')
axoplossingE.set_ylabel('y')
figoplossingE.savefig("oplossingE.pdf",bbox_inches='tight')
plt.show()
```
