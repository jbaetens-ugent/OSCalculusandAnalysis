### FreeFall_DF2

```python
plotname = "FreeFall_DF2.pdf"

import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, Function, Derivative, dsolve
from sympy.utilities.lambdify import lambdify

g=9.81
m=75
mu=5.25
t, v = np.meshgrid(np.linspace(0,30,15),np.linspace(0,120,14))
dt = 1
dv = g-mu*v/m
plt.quiver(t,v,2*dt/np.sqrt(dv**2+dt**2),dv/np.sqrt(dv**2+dt**2))

t = symbols('t')
v = Function('v')
v1 = Derivative(v(t), t)
sol = dsolve(m*v1 - m*g + mu*v(t), v(t), ics={v(0):0})
f = lambdify(t, sol.rhs, 'numpy')
t = np.linspace(0, 30, num=100)
plt.plot(t, f(t))
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')

plt.savefig(f"plots/{plotname}.pdf", pad_inches=0.1, transparent="True", bbox_inches='tight')
```

### FreeFall_DF3

```python
plotname = "FreeFall_DF3.pdf"

import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, Function, Derivative, dsolve
from sympy.utilities.lambdify import lambdify

g=9.81
m=75
mu=5.25
t, v = np.meshgrid(np.linspace(0,85,15),np.linspace(0,250,14))
dt = 1
dv = g-mu*v/m
plt.quiver(t,v,2*dt/np.sqrt(dv**2+dt**2),dv/np.sqrt(dv**2+dt**2))

t = symbols('t')
v = Function('v')
v1 = Derivative(v(t), t)
sol = dsolve(m*v1 - m*g + mu*v(t), v(t), ics={v(0):0})
f = lambdify(t, sol.rhs, 'numpy')
t = np.linspace(0, 85, num=100)
plt.plot(t, f(t), c='k', label="0")

t = symbols('t')
v = Function('v')
v1 = Derivative(v(t), t)
sol = dsolve(m*v1 - m*g + mu*v(t), v(t), ics={v(0):60})
f = lambdify(t, sol.rhs, 'numpy')
t = np.linspace(0, 85, num=100)
plt.plot(t, f(t), ":", c='k', label="60")

t = symbols('t')
v = Function('v')
v1 = Derivative(v(t), t)
sol = dsolve(m*v1 - m*g + mu*v(t), v(t), ics={v(0):120})
f = lambdify(t, sol.rhs, 'numpy')
t = np.linspace(0, 85, num=100)
plt.plot(t, f(t), "-.", c='k', label="120")

t = symbols('t')
v = Function('v')
v1 = Derivative(v(t), t)
sol = dsolve(m*v1 - m*g + mu*v(t), v(t), ics={v(0):180})
f = lambdify(t, sol.rhs, 'numpy')
t = np.linspace(0, 85, num=100)
plt.plot(t, f(t), ":", c="gray", label="180")

t = symbols('t')
v = Function('v')
v1 = Derivative(v(t), t)
sol = dsolve(m*v1 - m*g + mu*v(t), v(t), ics={v(0):240})
f = lambdify(t, sol.rhs, 'numpy')
t = np.linspace(0, 85, num=100)
plt.plot(t, f(t), c="gray", label="240")

plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.legend()

plt.savefig(f"plots/{plotname}.pdf", pad_inches=0.1, transparent="True", bbox_inches='tight')
```
